```python
# 1. Data -- data3d

#2. Scales
col_sc = bqplot.ColorScale(scheme="blues") 
x_sc = bqplot.OrdinalScale() # treating bins like bins of categorical data
y_sc = bqplot.OrdinalScale()

#3. Axis -- color & for x/y
ax_col = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
ax_x = bqplot.Axis(scale=x_sc,label = 'Year',tick_rotate=45,tick_style = {'text-anchor': 'start'},label_offset = '240')
ax_y = bqplot.Axis(scale=y_sc, orientation='vertical',side='left',label = 'Agencies',label_offset='-240')

# 4. Mark -- heatmap
heat_map = bqplot.GridHeatMap(color=pivot_values, 
                             row= pivot_df.index, column = pivot_df.columns,
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc},
                             interactions={'click':'select'},
                             anchor_style={'fill':'blue'})

# 5. interactions 
mySelectedLabel = ipywidgets.Label()
mySelectedLabel.value = 'Mean selected point here'
# interaction function -- what happens on selection:
def on_selected(change):
    #print(change)
    if len(change['owner'].selected) == 1:
        i,j = change['owner'].selected[0] 
#         print(i,j)
        v = pivot_values[i,j].mean() 
        mySelectedLabel.value = 'Mean YTD Gross salary = $'+str(round(v,2))
        a = [pivot_df.index[i]]
        b = [pivot_df.columns[j]]
        c = pay_data[pay_data['Year'].isin(b)]
        p_data_df = c[c['Agency'].isin(a)]
#         year_val = d.groupby('Year').agg(Total_Square_Footage = ('YTD Gross','sum'))        # update histogram based on selection
#         if len(p_data_df) > 1:
#             Bar.sample = p_data_df['YTD Gross']
#         Bar.y = p_data_df['YTD Gross'].values
#         if len(p_data_df) > 2:
        z = p_data_df.groupby('Employment Status').agg(YTD_Gross=('YTD Gross','mean'))
        if len(z)>1:
            pie.sizes=z
        p = p_data_df.groupby('Collective or Merit').agg(YTD_Gross=('YTD Gross','mean'))
#         if len(p>1):
#             Bar2.x = p.index
#             Bar2.y = p['YTD_Gross'].values
#         hist.sample = year_val
# link interaction function to my mark:
heat_map.observe(on_selected, 'selected')

# 6. Finally, put together figure
fig = bqplot.Figure(title="Heatmap for mean YTD Gross salary for each agency from 2011 to 2021",
                    marks=[heat_map], axes=[ax_col, ax_x, ax_y], 
                    fig_margin={'top':70, 'bottom':50, 'left':250, 'right':100})
fig.layout.width = "800px"
fig.layout.height = "1000px"
# display!
#fig
# myDashboard = ipywidgets.VBox([mySelectedLabel, fig])
# myDashboard
```

#### As we can see below, this is an interactive dashboard. The dashboard consists of a heat map and a pie chart. 
#### So before we begin, what is a heat map?
#### It is nothing but a chart which has small small squares inside it. The colour of the squares indicate how small or how large the value is. As we can see below, dark red values are smaller values and dark green values are bigger values.
#### So now that we understood what a heat map is, we can now have a look at the below dashboard. The heatmap in the dashboard consists of "Years" on the x-axis and "Agency" on the y-axis. Every value inside the heatmap represents mean of the annual salaries of all the employees in a particular Agency in that particular year. For example, if we click on the agency "State Police" in the year "2016", we can see that the mean gross annual salary for all the police personnal was 56,875.17 USD. Similarly we can click on other areas in the heat map to get the average salaries of all the employees in an agency and in a particular year.
#### Now coming to the other plot inside the dashboard i.e. the Pie plot. It is interactive and changes when you click on those tiny squares inside the heatmap. 
#### What? why? 
#### The Pie plot first tells us how many employees were employed in each agency. So when we click on each of the tiny square inside the heatmap, we select a particular year, for example when we selected the state police in 2016, we can see the number of employees across all the agencies in "2016". Isn't that cool? 

#### So to sum it all up, the Heatmap gives us the average salaries of all the employees in a particular agency and in a particular year. The Pie plot gives us the number of employees across all the agencies in the year selected in the Heatmap.
